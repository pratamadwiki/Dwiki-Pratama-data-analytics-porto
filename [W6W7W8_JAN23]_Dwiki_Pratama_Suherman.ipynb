##Check outlier

```python
# remove outliers function
def iqr(df,x):
    # setup the outlier parameter rules
    Q1 = df[x].quantile(0.25)
    Q3 = df[x].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    # placing the outlier to a variable
    filter_price_min = df[x]<lower_fence
    filter_price_max = df[x]>upper_fence

    # eliminate the outlier from the data frame
    df = df[~(filter_price_min|filter_price_max)]

    # print the outlier rules
    print('Q1:\n',Q1)
    print('\nQ3:\n',Q3)
    print('\nIQR:\n',IQR)
    print('\nLower Fence:\n',lower_fence)
    print('\nUpper Fence:\n',upper_fence)
    
    # visualize the outlier
    df.boxplot(column=[x],fontsize=10,rot=0,grid=False,figsize=(10,2),vert=False)
    return df
```

```python
#Check the Q1,Q3,IQR,UB & LB
#ainc = total buy
print('Outlier Detection "total":\n')
ainc_Q1 = df_merged["total"].quantile(0.25)
print("Q1 =",ainc_Q1)

ainc_Q3 = df_merged["total"].quantile(0.75)
print("Q3 =",ainc_Q3)

ainc_IQR = ainc_Q3 - ainc_Q1
print("IQR =",ainc_IQR)

ainc_LB = ainc_Q1-1.5*ainc_IQR
print("LB =",ainc_LB)

ainc_UB = ainc_Q3+1.5*ainc_IQR
print("UB =",ainc_UB)

#------------------------------------------
#lamnt =total sell
print('Outlier Detection "sell":\n')
lamnt_Q1 = df_merged["sell"].quantile(0.25)
print("Q1 =",lamnt_Q1)

lamnt_Q3 = df_merged["sell"].quantile(0.75)
print("Q3 =",lamnt_Q3)

lamnt_IQR = lamnt_Q3 - lamnt_Q1
print("IQR =",lamnt_IQR)

lamnt_LB = lamnt_Q1-1.5*lamnt_IQR
print("LB =",lamnt_LB)

lamnt_UB = lamnt_Q3+1.5*lamnt_IQR
print("UB =",lamnt_UB)
```

```python
#Removing/Excluding the outlier and assign it on new variable "df_clean"
df_clean_annual_inc = df_merged[~((df_merged["total"] < ainc_LB) | (df_merged["total"] > ainc_UB))]
df_clean = df_clean_annual_inc[~((df_clean_annual_inc["sell"] < lamnt_LB) | (df_clean_annual_inc["sell"] > lamnt_UB))]
df_clean.shape
```

```python
sns.scatterplot (data = rs_scale_df, x='user_age', y='total')
```

##cluster

###elbow method

```python
distortions = []
K = range(1,11)
for k in K:
    kmeanModel = cluster.KMeans(n_clusters=k)
    kmeanModel.fit(rs_scale_df)
    distortions.append(kmeanModel.inertia_)
    print(k, kmeanModel.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

###silhoulette analysis

```python
# Silhouette score plot
K = range(2,11)
max_K = max(K)
fig, ax = plt.subplots(int(np.ceil(max_K/2)), 2, figsize = (15,30))

for n_clusters in K:
  kmeanModel = cluster.KMeans(n_clusters)

  q, mod = divmod(n_clusters,2)
  sil = SilhouetteVisualizer(kmeanModel, is_fitted = False, ax = ax[q-1][mod])
  sil.fit(rs_scale_df)
  sil.finalize()
  print(f"For k={n_clusters}, the average silhouette score is {sil.silhouette_score_}")
```

based on elbow method and silhouette analysis above, we can shoose cluster=4
